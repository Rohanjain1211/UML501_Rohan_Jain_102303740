```python
import pandas as pd
import numpy as np
```

```python
df=pd.read_csv('/content/AWCustomers.csv')
df
```

```python
df.columns
```

```python
cols=['CustomerID','MiddleName','CountryRegionName','BirthDate','Occupation','Gender','MaritalStatus','HomeOwnerFlag','NumberCarsOwned','NumberChildrenAtHome','YearlyIncome','LastUpdated']
df1=df[cols]
df1
```

```python
df1['MiddleName']=df1["MiddleName"].fillna('NA')
df1
```

```python

df1.drop(columns=['CustomerID','MiddleName'],inplace=True,axis=1)
df1
```

```python
numeric_cols = ['YearlyIncome', 'NumberCarsOwned', 'NumberChildrenAtHome']

df1['BirthDate'] = pd.to_datetime(df1['BirthDate']).dt.year

df1['Gender'] = df1['Gender'].map({'M': 1, 'F': 0})
df1['MaritalStatus']=df1['MaritalStatus'].map({'M': 1, 'S': 0})

df1
```

```python
from sklearn.preprocessing import MinMaxScaler,StandardScaler,OneHotEncoder,KBinsDiscretizer
numeric_cols = ['YearlyIncome', 'NumberCarsOwned', 'NumberChildrenAtHome']
scaler=MinMaxScaler()
df1[numeric_cols]=scaler.fit_transform(df1[numeric_cols])
df1
```

```python
from sklearn.preprocessing import LabelEncoder

le=LabelEncoder()

df1['Occupation']=le.fit_transform(df1['Occupation'])
df1
```

```python
ohe=OneHotEncoder()

country_encoded = ohe.fit_transform(df1[['CountryRegionName']]).toarray()

# Create a DataFrame from the one-hot encoded data
country_df = pd.DataFrame(country_encoded, columns=ohe.get_feature_names_out(['CountryRegionName']), index=df1.index)

# Drop the original column and concatenate the new one-hot encoded columns
df1 = df1.drop('CountryRegionName', axis=1).join(country_df)

df1
```
